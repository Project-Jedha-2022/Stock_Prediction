```python
import pandas as pd
import numpy as np
```

```python
social = pd.read_csv('/content/ubisoft_reddit_sentiment_analysis.csv')
technical = pd.read_csv('/content/ubisoft_technical_dataset_1d.csv')
```

```python
# Traiter Social jour + 1
date_1 = []

for i in range(0, social.shape[0]-1):
    # get row contents as series using iloc{]
    # and index position of row
    date_1.append(social.iloc[i+1]['date'])

date_1.append(np.nan)

social['date_1'] = date_1
social["date_1"] =  pd.to_datetime(social["date_1"], format="%Y-%m-%d")
```

```python
social
```

```python
social["date"] =  pd.to_datetime(social["date"], format="%Y-%m-%d")
technical["date"] =  pd.to_datetime(technical["date"], format="%Y-%m-%d")
```

```python
social
```

```python
technical
```

```python
dataset = technical.merge(social, left_on='date', right_on='date_1', how='inner')
dataset
```

```python
dataset.drop(['date_y', 'date_1'], axis=1, inplace=True)
dataset.rename(columns={"date_x": "date"}, inplace=True)
```

```python
dataset.to_csv('dataset_social_technical_1d.csv', index=False)
```
