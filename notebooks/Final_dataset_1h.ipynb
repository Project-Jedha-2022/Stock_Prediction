# Conversion des heures

```python
social["date"] =  pd.to_datetime(social["date"], format="%Y-%m-%d %H:%M:%S")
```

```python
social['date'] = social['date'].dt.strftime('%Y-%m-%d %H:00:00')
social
```

```python
social = social.groupby('date').mean().reset_index()
```

```python
social
```

```python
dataset = technical.merge(social, left_on='date', right_on='date', how='inner')
dataset
```

```python
dataset.to_csv('dataset_social_technical_1h.csv', index=False)
```
